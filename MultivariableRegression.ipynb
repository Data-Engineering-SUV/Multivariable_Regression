#Analyzing the OVERFITTING case of Linear Regression Algorithm upto 2 Degree polynomial

```python
import numpy as np
import pandas as pd

# retreving boston dataset
df = pd.read_csv('/content/sample_data/BostonHousing.csv')
df_X = df.iloc[:,:-1]

# Original Input & Output data
X = df_X.values
Y = df.iloc[:,-1].values
print("Shape of the input data: ",X.shape)
print("Shape of the output data: ",Y.shape)
```

```python
df_X.head()
```

```python
All_columns = df_X.columns.values
# Addition of new feature upto 2 degree polinomial (Overfitting Case)

# 1st Squre of every feature
for i in All_columns:
  df_X[str(i)+ '_' +str(i)] = df_X[i]**2
print(df_X.shape)
df_X.columns.values
```

```python
# 2nd adding New feature by multipling with other feature
# 13C2 + 13 + 13 = 104 (Total should be 104 feature)
for i in All_columns:
  for j in All_columns:
    if i == j:
      break
    else:
      bina = df_X[i] * df_X[j]
      df_X[str(j) + '_' + str(i)] = bina
print(df_X.shape)
df_X.columns.values
```

```python
# After adding all feature up to 2 Degree polynomial
df_X.head()
```

###Updated input data with all 2 degree polynomial

```python
# Original input & output feature is X and Y
# After adding multivariables input & output is X2 and Y
X2 = df_X.values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
```

```python
# Putting Underfitting and Overfitting case in two algorithm

X1_train, X1_test, Y1_train, Y1_test = model_selection.train_test_split(X,Y,random_state = 0)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(X2,Y,random_state = 0)
algo_one = LinearRegression()
algo_two = LinearRegression()
algo_one.fit(X1_train,Y1_train)
algo_two.fit(X2_train,Y2_train)
```

```python
train_score1 = algo_one.score(X1_train, Y1_train)
test_score1 = algo_one.score(X1_test, Y1_test)
print("Algorithm One score>> ","Training Score:",train_score1, ",","Testing Score:", test_score1)

train_score2 = algo_two.score(X2_train, Y2_train)
test_score2 = algo_two.score(X2_test, Y2_test)
print("Algorithm Two score>> ","Training Score:",train_score2, ",","Testing Score:", test_score2)
```
